```python
import pandas as pd
import numpy as np
import os
import time
```

```python
def read_return_data(name):
    filename="./data/csi300_alpha_data/"+name+'_csi300.csv'
    data=pd.read_csv(filename)
    data=pd.melt(data,id_vars=['datetime'])
    data.rename(columns={0:"datetime", 'variable':"tickers", 'value':name},inplace=True)
    data['datetime']=pd.to_datetime(data['datetime'])
    data=data.fillna(method='ffill')
    return data
```

```python
filelist = ['eps','roa','pb','pe','dividend',
            'Alpha1','Alpha4','Alpha5','Alpha13','Alpha15','Alpha16','Alpha18',
            'Alpha25','Alpha26','Alpha33','Alpha35','Alpha37','Alpha57','Alpha84',
            'Alpha101','Alpha103']
```

```python
merge_data=read_return_data('return')
merge_data.head()
```

```python
for alpha in filelist:
    print "Now merge alpha: ", alpha
    data=read_return_data(alpha)
    merge_data=pd.merge(merge_data,data,on=['datetime','tickers'],how='inner')
```

```python
merge_data.to_csv("merge_data.csv",index=0)
print "Merge data finished !"
merge_data.head()
```
